## Player summary

```python
player_fx = data.get_player_summary(355, 'fixtures')
player_fx
```

```python
def style_background_player_fdr(cell_value):

    bg = 'background-color:'

    if cell_value == 1:
        return f'{bg} darkgreen;'
    elif cell_value == 2:
        return f'{bg} green;'
    elif cell_value == 3:
        return f'{bg} grey;'
    elif cell_value == 4:
        return f'{bg} orange;'
    elif cell_value == 5:
        return f'{bg} darkred;'
    else:
        return ''
    
player_fx.head(10).T.style.applymap(style_background_player_fdr)
```

## Fixtures
Try to create a fixtures matrix, showing each team's upcoming fixtures and their difficulty

```python
fx = data.get_fixtures_matrix()
fx
```

```python
def style_background_team_fdr(cell_value):

    bg = 'background-color:'
    team_str = 0

    if cell_value != '':
        team_str = int(cell_value[-1])

    if team_str == 1:
        return f'{bg} darkgreen;'
    elif team_str == 2:
        return f'{bg} green;'
    elif team_str == 3:
        return f'{bg} grey;'
    elif team_str == 4:
        return f'{bg} orange;'
    elif team_str == 5:
        return f'{bg} darkred;'
    else:
        return ''

fx.style.applymap(style_background_team_fdr)
```
