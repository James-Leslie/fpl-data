# Using `fpl_data.transform` module

```python
import pandas as pd
from fpl_data.transform import FplApiDataTransformed

pd.set_option('display.max_columns', None)
```

Load and transform data

```python
data = FplApiDataTransformed()
```

## Players

```python
data.players.sort_values('Pts', ascending=False)
```

## Fixtures

```python
data.fixtures[0]
```

```python
team_names = data.teams[['team']]

fixtures = pd.json_normalize(
    data.fixtures
).merge(
    # join to team names (home)
    team_names,
    left_on='team_h',
    right_on='team_id',
    suffixes=[None, '_home']
).merge(
    # join to team names (away)
    team_names,
    left_on='team_a',
    right_on='team_id',
    suffixes=[None, '_away']
).rename(
    columns={
        'id': 'fixture_id',
        'event': 'GW',
        'team': 'team_home'
    }
).drop(
    ['code', 'finished_provisional', 'kickoff_time', 'minutes',
     'provisional_start_time', 'started', 'stats', 'pulse_id'],
    axis=1
)
fixtures
```

```python
# filter between start_gw and end_gw
start_gw = data.next_gw

end_gw = start_gw + 8

fixtures = fixtures[
    (fixtures['GW'] >= start_gw) & (fixtures['GW'] <= end_gw)]

fixtures
```

```python
fx_ratings, fx_team_names = data.get_fixtures_matrix()
fx_ratings
```

```python
str_lookup = {}
for i, r in data.teams.iterrows():
    str_lookup[r['team']] = r['strength']
```

```python
def style_background(cell_value):

    bg = 'background-color:'

    try:
        team_str = str_lookup[cell_value[:3]]
    except (KeyError, TypeError) as error:
        team_str = cell_value

    if team_str == 1:
        return f'{bg} darkgreen;'
    elif team_str == 2:
        return f'{bg} green;'
    elif team_str == 3:
        return f'{bg} grey;'
    elif team_str == 4:
        return f'{bg} orange;'
    elif team_str == 5:
        return f'{bg} darkred;'
    else:
        return ''
```

```python
fx_ratings.style.applymap(style_background)
```

```python
fx_team_names.style.applymap(style_background)
```
