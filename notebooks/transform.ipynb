# Using `fpl_data.transform` module

```python
import pandas as pd
from fpl_data.transform import FplApiDataTransformed

pd.set_option('display.max_columns', None)
```

Load and transform data

```python
data = FplApiDataTransformed()
```

## Players

```python
# get a list of all players as a dataframe
data.players_df.sort_values('Pts', ascending=False)
```

## Player summary
The `.get_player_summary()` method can be used to get a player's past or future fixtures

```python
# get the upcoming fixtures for a player
player_fx = data.get_player_summary(355, 'fixtures')
player_fx.head()
```

We can apply some styling to the result above to get a more graphical representation of the upcoming fixtures' difficulty

```python
def style_background_player_fdr(cell_value):

    bg = 'background-color:'

    if cell_value == 1:
        return f'{bg} darkgreen;'
    elif cell_value == 2:
        return f'{bg} green;'
    elif cell_value == 3:
        return f'{bg} grey;'
    elif cell_value == 4:
        return f'{bg} orange;'
    elif cell_value == 5:
        return f'{bg} darkred;'
    else:
        return ''

# ToDo: make this part of the package
player_fx.head(10).T.style.map(style_background_player_fdr)
```

## Fixtures
Try to create a fixtures matrix, showing each team's upcoming fixtures and their difficulty

```python
fx = data.get_fixtures_matrix()
fx
```

```python
def style_background_team_fdr(cell_value):

    bg = 'background-color:'
    team_str = 0

    if cell_value != '':
        team_str = int(cell_value[-1])

    if team_str == 1:
        return f'{bg} darkgreen;'
    elif team_str == 2:
        return f'{bg} green;'
    elif team_str == 3:
        return f'{bg} grey;'
    elif team_str == 4:
        return f'{bg} orange;'
    elif team_str == 5:
        return f'{bg} darkred;'
    else:
        return ''

fx.style.map(style_background_team_fdr)
```
